Scatter plots are the graphs that present the relationship between two variables in a dataset.

```python
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
from matplotlib import style
style.use("seaborn")
```

```python
x = [9,4,6,1,7,2,6,3,1,7]
y = [1,5,6,2,7,9,3,1,5,9]
```

```python
plt.scatter(x,y) #show with object
```

```python
plt.scatter(x,y); #show without object
```

```python
plt.scatter(x,y)
plt.title("Scatter Plot"); #to give title
```

```python
plt.scatter(x,y)
plt.ylabel("Y axis =>") #to give 'y' axis
plt.xlabel("X axis =>"); #to give "x" axis
plt.show() #to show if not work
```

```python
plt.scatter(x,y,s=200); #to increase scatter size
```

```python
plt.scatter(x,y,c="green"); #to change scatter color
```

```python
plt.scatter(x,y,marker="X"); #to change marker style
```

```python
plt.scatter(x,y,c="yellow", edgecolors="black", linewidths=1); #to style scatter
```

```python
x = [9,4,6,1,7,2,6,3,1,7]
y = [1,5,6,2,7,9,3,1,5,9]
colors = [4,5,6,1,4,7,9,8,5,2]
plt.scatter(x,y, c=colors); #to give different colors
```

```python
x = [9,4,6,1,7,2,6,3,1,7]
y = [1,5,6,2,7,9,3,1,5,9]
colors = [4,5,6,1,4,7,9,8,5,2]
plt.scatter(x,y, c=colors,cmap="Accent", edgecolors="green"); #to give different color range
```

```python
x = [9,4,6,1,7,2,6,3,1,7]
y = [1,5,6,2,7,9,3,1,5,9]
colors = [4,5,6,1,4,7,9,8,5,2]
plt.scatter(x,y, c=colors); #to give different colors
"""to set color bar"""
cbar = plt.colorbar()
cbar.set_label("Random")
plt.show()
```

```python
x = [9,4,6,1,7,2,6,3,1,7]
y = [1,5,6,2,7,9,3,1,5,9]
colors = [4,5,6,1,4,7,9,8,5,2]
"""to give different-different size"""
sizes = [500, 20, 332, 651, 851, 493, 211, 291, 333, 421]
plt.scatter(x,y, c=colors, s=sizes); #to give different colors
```
