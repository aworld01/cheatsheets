```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
l = [9,10,22,23,24,25,26,27,28,29,30,40,41]
```

```python
plt.boxplot(l)
plt.show()
```

```python
"""to show plot horizontally"""
plt.boxplot(l, vert=False) #to show plot horizontally (default is True)
plt.show()
```

```python
plt.boxplot(l, vert=False, showmeans=True) #to see means value (default is False)
plt.show()
```

```python
plt.boxplot(l, vert=False, sym="b+") #to change outlayers into blue and plus
plt.show()
```

```python
plt.boxplot(l, vert=False, sym="b+", notch=1) #to give notch to midiun
plt.show()
```

```python
plt.boxplot(l, vert=False, sym="") #to remove outlayers
plt.show()
```

```python
"""multi designing outlayers"""
blue_circles = dict(markerfacecolor="blue", marker="o")
plt.boxplot(l, flierprops=blue_circles)
plt.show()
```

```python
blue_circles = dict(markerfacecolor="blue", marker="o", markeredgecolor="yellow")
plt.boxplot(l, flierprops=blue_circles)
plt.show()
```

```python
"""multi design mean shape"""
mean_shape = dict(markerfacecolor="red", marker="D", markeredgecolor="blue")
plt.boxplot(l, showmeans=True, meanprops=mean_shape)
plt.show()
```
