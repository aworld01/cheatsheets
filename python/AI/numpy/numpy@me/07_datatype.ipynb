```python
import numpy as np
```

```python
a = np.arange(10)
a
```

```python
a.dtype
```

```python
a = np.arange(10, dtype="float64") #you can explicitly specify which datatype you want to create
print(a)
a.dtype
```

```python
z = np.zeros((3,4)) #the default datatype is float for zeros and ones function
z.dtype
```

```python
"""complex datatype"""
d = np.array([1+2j, 2+4j])
print(d)
d.dtype
```

```python
"""boolean datatype"""
d = np.array([True, False, True, False])
print(d)
print(d.dtype)
```
